```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

# Define the function for reading and processing the data
def read_and_process(filename, rotation_angle=0):
    # Read the data from the CSV file
    data = np.genfromtxt(filename, delimiter=',')
    
    # Rotate the image by the given angle
    rotated_image = cv2.rotate(data, rotation_angle)
    
    # Normalize and handle missing or zero values
    return rotated_image

# Initialize empty arrays to store concentration values for each element
concentration_dict = {
    'Ni': [],
    'Al': [],
    'Co': [],
    'Cr': [],
    'Fe': [],
    'Ti': [],
    'V': [],
    'O': [],
    'C': [],
    'N': []
}

# Define the pixel range (px, py) for the analysis
px, py = 500, 500
x = np.linspace(0, 70, px)

# List of filenames for each element (assuming the file names are as provided in the original code)
filenames = [
    'data001.csv', 'data002.csv', 'data003.csv', 'data004.csv', 
    'data005.csv', 'data006.csv', 'data007.csv', 'data008.csv',
    'data009.csv', 'data010.csv'
]

# Loop through the filenames, read the data, and calculate the mean for each element
for i, filename in enumerate(filenames):
    rotated_image = read_and_process(filename)
    
    # Calculate the mean value for each row (ignoring zeros)
    concentration = []
    for i in range(px):
        row_data = rotated_image[i, :]
        valid_data = row_data[row_data != 0]  # Exclude zero values
        mean_value = np.mean(valid_data) if len(valid_data) > 0 else 0
        concentration.append(mean_value)
    
    # Map the concentrations to the corresponding elements
    element = list(concentration_dict.keys())[i]
    concentration_dict[element] = concentration

# Plot the line profiles for all elements
plt.figure(figsize=(10, 6))
for element, concentration in concentration_dict.items():
    plt.plot(x, concentration, label=element)

# Customize plot
plt.xlabel('Depth (μm)', fontsize=22)
plt.ylabel('Concentration (at%)', fontsize=22)
plt.legend(loc='northeastoutside', fontsize=16)
plt.tick_params(axis='both', labelsize=20)

# Save the plot
plt.savefig('Line_profile_all.png', dpi=300)
plt.show()
```

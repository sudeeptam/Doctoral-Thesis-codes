```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Font parameters (slightly adjusted)
font_params = {'fontsize': 20, 'fontweight': 'bold', 'family': 'sans-serif'}
tick_params = {'labelsize': 14}

# Data
oxide_data = {
    "B": [32.95, 74.54, 65.21],
    "B-10Co": [15.10, 49.91, 24.35],
    "B-10Co-5Fe": [10.74, 17.97, 47.49],
    "B-10Co-11.5Fe": [25.27, 18.58, 13.86],
    "B-20Co": [15.47, 60.21, 24.07],
    "B-30Co": [34.05, 22.52, 16.63],
    "B-30Co-2V": [15.38, 71.85, 21.43]
}
oxide_std_dev = {
    "B": [8.85, 9.99, 8.32],
    "B-10Co": [4.94, 4.67, 5.42],
    "B-10Co-5Fe": [3.49, 1.83, 8.47],
    "B-10Co-11.5Fe": [2.05, 2.32, 1.62],
    "B-20Co": [3.78, 3.27, 2.87],
    "B-30Co": [9.10, 2.32, 4.46],
    "B-30Co-2V": [2.63, 4.35, 4.42]
}

alloys = list(oxide_data.keys())
temps = [900, 950, 1000]

X, Y = np.meshgrid(range(len(alloys)), temps)
Z = np.array([oxide_data[alloy] for alloy in alloys]).T
errors = np.array([oxide_std_dev[alloy] for alloy in alloys]).T

xpos = X.flatten()
ypos = Y.flatten()
zpos = np.zeros_like(xpos)
dx = dy = 0.6
dz = Z.flatten()
err = errors.flatten()

# Replotting with axis labels removed and using manual layout adjustments

fig = plt.figure(figsize=(18, 10))
fig.subplots_adjust(left=0.1, right=0.9, bottom=0.25, top=0.95)
ax = fig.add_subplot(111, projection='3d')

# Plot bars
ax.bar3d(xpos, ypos, zpos, dx, dy, dz, color='grey', edgecolor='black')

# Add error bars
for i in range(len(xpos)):
    ax.plot([xpos[i]+dx/2, xpos[i]+dx/2], [ypos[i]+dy/2, ypos[i]+dy/2],
            [dz[i], dz[i] + err[i]], color='black', linewidth=2)

# Axis and ticks (labels removed)
ax.set_xticks(np.arange(len(alloys)) + 0.3)
ax.set_xticklabels(alloys, rotation=30, ha='right', **font_params)
ax.set_yticks(temps)
ax.set_yticklabels([f"{t}°C" for t in temps], **font_params)
ax.set_zlabel("")
ax.set_ylabel("")
ax.set_xlabel("")
ax.tick_params(axis='both', which='major', **tick_params)

# Spine styling
for spine in [ax.xaxis.line, ax.yaxis.line, ax.zaxis.line]:
    spine.set_linewidth(2)

plt.show()
```

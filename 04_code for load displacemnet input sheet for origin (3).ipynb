```python
# -*- coding: utf-8 -*-
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from math import pi

# ---------- CONFIG ----------
FILEPATH = r"H:\compression\Sudeepta@suwas\to copy\co30v2\B-30Co-2V 570 4_cleaned.xlsx"
DIAMETER_MM = 2.5        # cylindrical specimen diameter (mm)
GAUGE_LEN_MM = 4.3       # gauge length (mm)
FIT_LOW, FIT_HIGH = 5.0, 7.0   # fit window for "elastic" slope (in % strain)
OFFSET = 0.2             # 0.2% offset
# ----------------------------

# Read Excel and auto-detect columns
df = pd.read_excel(FILEPATH)
load_col = next(c for c in df.columns if "load" in c.lower())
stroke_col = next(c for c in df.columns if ("stroke" in c.lower()) or ("disp" in c.lower()))
df = df[[load_col, stroke_col]].copy()
df.columns = ["Load_kN", "Stroke_mm"]
df = df.apply(pd.to_numeric, errors="coerce").dropna()

# Stress–strain
AREA_MM2 = pi * (DIAMETER_MM/2)**2
df["EngStress_MPa"] = df["Load_kN"] * (1000.0 / AREA_MM2)               # MPa
df["EngStrain_%"]   = (df["Stroke_mm"] / GAUGE_LEN_MM) * 100.0          # %

# Sort by strain (helps later)
df = df.sort_values("EngStrain_%").reset_index(drop=True)

# --- Fit slope in the chosen window (5–7% by default) ---
subset = df[(df["EngStrain_%"] >= FIT_LOW) & (df["EngStrain_%"] <= FIT_HIGH)].copy()
if len(subset) < 3:
    # Gentle fallback: widen window a bit if too few points
    widen = 0.5
    subset = df[(df["EngStrain_%"] >= FIT_LOW - widen) & (df["EngStrain_%"] <= FIT_HIGH + widen)]
    if len(subset) < 3:
        raise RuntimeError("Not enough points in/near 5–7% to fit slope. Consider adjusting the window.")

x_fit = subset["EngStrain_%"].values
y_fit = subset["EngStress_MPa"].values
slope, intercept = np.polyfit(x_fit, y_fit, 1)   # stress = slope * strain(%) + intercept

# --- Back-extrapolate 0–5% with this fitted line ---
mask_back = df["EngStrain_%"] <= 5.0
df.loc[mask_back, "EngStress_MPa"] = slope * df.loc[mask_back, "EngStrain_%"] + intercept

# --- Shift curve so stress = 0 is at strain = 0 ---
# In the (0–5%) region, curve is linear: stress = slope*strain + intercept
# Zero-stress crossing of that line:
x_cross = -intercept / slope
df["EngStrain_Shifted_%"] = df["EngStrain_%"] - x_cross

# --- 0.2% offset line (parallel to elastic slope, shifted by +0.2% strain) ---
# y_off = slope * (x_shifted - OFFSET)
x = df["EngStrain_Shifted_%"].values
y = df["EngStress_MPa"].values
y_off = slope * (x - OFFSET)

# Intersection: solve y - y_off = 0
y_diff = y - y_off
sign_changes = np.where(np.diff(np.sign(y_diff)) != 0)[0]

if len(sign_changes) == 0:
    # Fallback: use nearest approach
    i = int(np.argmin(np.abs(y_diff)))
    x_yield = x[i]
    y_yield = y[i]
else:
    i = sign_changes[0]
    x1, x2 = x[i],   x[i+1]
    d1, d2 = y_diff[i], y_diff[i+1]
    # Linear interpolation for crossing
    t = -d1 / (d2 - d1)
    x_yield = x1 + t * (x2 - x1)
    y_yield = np.interp(x_yield, x, y)

print(f"Slope (5–7%): {slope:.2f} MPa/%")
print(f"Zero-stress crossing (pre-shift): {x_cross:.4f}% strain")
print(f"0.2% offset yield stress: {y_yield:.2f} MPa at {x_yield:.3f}% strain (shifted scale)")

# --- Plot with your style ---
plt.figure(figsize=(6,6))
plt.plot(x, y, color="black", linewidth=2.5, label="Corrected curve")
# draw the 0.2% offset line over the same x-range
x_line = np.linspace(max(0, x.min()), x.max(), 400)
plt.plot(x_line, slope*(x_line - OFFSET), "--", linewidth=2, label="0.2% offset line")
plt.scatter([x_yield], [y_yield], s=60, zorder=5, label=f"0.2% offset = {y_yield:.1f} MPa")

# Style: Candara 22 bold, grid off, all spines on, thick ticks
plt.xlabel("Engineering Strain (%)", fontname="Candara", fontsize=22, fontweight="bold")
plt.ylabel("Engineering Stress (MPa)", fontname="Candara", fontsize=22, fontweight="bold")
plt.title("B-30Co-2V 570°C", fontname="Candara", fontsize=22, fontweight="bold")

plt.grid(False)
ax = plt.gca()
for spine in ax.spines.values():
    spine.set_visible(True)
    spine.set_linewidth(2)
plt.tick_params(axis='both', direction='in', length=8, width=2, labelsize=18)

plt.xlim(0, max(0.05, x.max()*1.05))
plt.ylim(0, y.max()*1.05)
plt.legend(fontsize=12)
plt.tight_layout()
plt.show()

# Save data (optional)
out_csv = FILEPATH.replace(".xlsx", "_stress_strain_corrected_and_shifted.csv")
df.to_csv(out_csv, index=False)
print(f"Saved: {out_csv}")
```
